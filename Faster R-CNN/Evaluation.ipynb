```python
def sort_list(list1, list2): 
  
    zipped_pairs = zip(list2, list1) 
  
    z = [x for _, x in sorted(zipped_pairs)] 
      
    return z 
```

```python
new_labels = sort_list(labels, AP_Resnet)
new_AP_Resnet = sorted(AP_Resnet)
new_AP_Desnet = sort_list(AP_Desnet, AP_Resnet)
new_AP_MobileNet = sort_list(AP_MobileNet, AP_Resnet)
new_AP_YOLO = sort_list(AP_YOLO, AP_Resnet)
new_AP_VGG = sort_list(AP_VGG, AP_Resnet)
new_AP_YOLO_VGG = sort_list(AP_YOLO_VGG, AP_Resnet)
```
